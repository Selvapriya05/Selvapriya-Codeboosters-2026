```python
import pandas as pd
```

```python
df=pd.read_csv('/content/student_performance.csv')
```

```python
print(df)
```

```python
print(f"Total Students: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Department Count: {df['department'].nunique()}")
```

```python
print(df['department'].value_counts())
```

```python
print(f"Highest Math Score:{df['math_score'].max()}")
print(f"Lowest Math Score:{df['math_score'].min()}")
print(f"Average Math Score:{df['math_score'].mean().round(2)}")
print(f"Average Programming Score:{df['programming_score'].mean().round(2)}")
print(f"Average Attendance Percentage:{df['attendance_percentage'].mean().round(2)}%")
```

```python
df['total_score']=(df['math_score']+df['science_score']+df['english_score']+df['programming_score'])
gender_score = df.groupby('gender')['total_score'].mean().round(2)
print(gender_score)
```

```python
top5=df.sort_values('total_score',ascending=False).head(5)
for index, row in top5.iterrows():
    print(f"{row['name']:<20}  {row['department']:<18}  Score:{row['total_score']}")
```

```python
low_attendance=df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_attendance)>0:
  print(low_attendance.to_string(index=False))
else:
  print("All Students have attendance above 75%")
print("Report Complete")
```

```python
electronics_students = df[df['department'] == 'Electronics']
print("1. Electronics Department Students:",
      len(electronics_students))
```

```python
female_students = df[df['gender'] == 'Female']
print("2. Average Attendance of Female Students:",
      round(female_students['attendance_percentage'].mean(), 2))
```

```python
lowest_programming = df.loc[df['programming_score'].idxmin()]
print("3. Student with Lowest Programming Score:",
      lowest_programming['name'])
```

```python
math_above_80 = df[df['math_score'] > 80]
print("4. Students Scored Above 80 in Math:",
      len(math_above_80))
```

```python
# Create total_score column
df['total_score']=(df['math_score']+df['science_score']+df['english_score']+df['programming_score'])
# Students with attendance above 90%
high_attendance = df[df['attendance_percentage'] > 90]
# Average total score
print("Average Total Score:",
      high_attendance['total_score'].mean())
```
