```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

print("All libraries imported successfully!")
print(f"pandas version: {pd.__version__}")
print(f"sqlite3 version: {sqlite3.sqlite_version}")

# Load dataset
df = pd.read_csv('student_performance.csv')

print(f"Dataset loaded: {df.shape[0]} students, {df.shape[1]} columns")
print(f"Columns: {df.columns.tolist()}")

# Create database connection
conn = sqlite3.connect('college.db')
cursor = conn.cursor()

# Store dataframe into SQL table
df.to_sql(
    'students',
    conn,
    if_exists='replace',
    index=False
)

# Count rows
cursor.execute("SELECT COUNT(*) FROM students")
count = cursor.fetchone()[0]

print(f"Database 'college.db' created successfully!")
print(f"Table 'students' has {count} rows")

# Helper function
def run_query(sql, description="student"):

    if description:
        print(f"\n{'='*55}")
        print(f" {description}")
        print(f"{'='*55}")

    result = pd.read_sql_query(sql, conn)

    return result

# Query 1
query1 = """
SELECT student_id, name, department, math_score, attendance_percentage
FROM students
LIMIT 10;
"""

result1 = run_query(query1, "Query 1: First 10 Students")

# Query 2
query2 = """
SELECT student_id, name
FROM students
LIMIT 5;
"""

result2 = run_query(query2, "Query 2: First 5 Students")

# Query 3
query3 = """
SELECT name, math_score
FROM students
WHERE math_score > 90
ORDER BY math_score DESC
LIMIT 5;
"""

result3 = run_query(query3, "Query 3: Students with Math Score > 90")

# Query 4
query4 = """
SELECT name, math_score, programming_score, attendance_percentage
FROM students
WHERE math_score > 90
ORDER BY math_score DESC
LIMIT 5;
"""

result4 = run_query(query4, "Query 4: High Scoring Students")

# Query 5
query5 = """
SELECT name, department, attendance_percentage
FROM students
WHERE attendance_percentage > 93
AND department != 'Mechanical'
ORDER BY attendance_percentage DESC
LIMIT 3;
"""

result5 = run_query(query5, "Query 5: Attendance Analysis")

# Query 6
query6 = """
SELECT
    department,
    COUNT(*) AS num_students,
    ROUND(AVG(math_score),2) AS avg_math,
    ROUND(AVG(science_score),2) AS avg_science,
    ROUND(AVG(programming_score),2) AS avg_programming,
    ROUND(AVG(attendance_percentage),2) AS avg_attendance
FROM students
GROUP BY department
ORDER BY avg_math DESC;
"""

result6 = run_query(query6, "Query 6: Department Analysis")

# Query 7
query7 = """
SELECT
    department,
    ROUND(AVG(math_score),2) AS avg_math,
    COUNT(*) AS students
FROM students
GROUP BY department
HAVING AVG(math_score) > 70
ORDER BY avg_math DESC;
"""

result7 = run_query(query7, "Query 7: Departments with Avg Math > 70")

# Query 8
query8 = """
SELECT
    gender,
    COUNT(*) AS num_students,
    ROUND(AVG(math_score),2) AS avg_math,
    ROUND(AVG(programming_score),2) AS avg_programming,
    ROUND(AVG(attendance_percentage),2) AS avg_attendance
FROM students
GROUP BY gender;
"""

result8 = run_query(query8, "Query 8: Gender Wise Analysis")

# Display results
print(result1)
print(result2)
print(result3)
print(result4)
print(result5)
print(result6)
print(result7)
print(result8)
```
