```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("student_performance.csv")

conn = sqlite3.connect("students.db")

df.to_sql("students", conn, if_exists="replace", index=False)

print("Database created successfully")
```

```python
def get_schema(conn):
    cursor = conn.cursor()
    cursor.execute("PRAGMA table_info(students)")
    schema = cursor.fetchall()

    schema_text = "Table students columns:\n"
    for col in schema:
        schema_text += f"- {col[1]} ({col[2]})\n"

    return schema_text
```

```python
def generate_sql(user_question, schema_text):

    prompt = f"""
You are an SQL expert.

Schema:
{schema_text}

Question:
{user_question}

Rules:
- Only return SQL query
- Table name: students
"""

    # SIMPLE MOCK (replace with Groq/OpenAI later)
    sql = "SELECT * FROM students LIMIT 5"

    return sql
```

```python
def execute_sql(sql, conn):
    try:
        df = pd.read_sql_query(sql, conn)
        return df, None
    except Exception as e:
        return None, str(e)
```

```python
def generate_answer(df):
    return f"Based on results, top students are:\n\n{df.to_string(index=False)}"
```

```python
def text_to_sql_agent(question, conn):

    print("\nQUESTION:", question)
    print("="*60)

    # STEP 1
    schema = get_schema(conn)

    # STEP 2
    sql = generate_sql(question, schema)

    print("\nGenerated SQL:")
    print(sql)

    # STEP 3
    result, error = execute_sql(sql, conn)

    if error:
        print("ERROR:", error)
        return

    # STEP 4
    print("\nRESULT TABLE:")
    print(result)

    # STEP 5
    print("\nNATURAL LANGUAGE ANSWER:")
    print(generate_answer(result))
```

```python
text_to_sql_agent(
    "Show top 5 students in Programming",
    conn
)
```
